### Visualizations

Pandas simplifies data visualization with its `.plot()` method, which acts as a convenient wrapper around matplotlib, Python's go-to plotting library.

Creating a basic plot for a column is as effortless as `df['Close'].plot()`. As we progress, we'll explore more advanced visualizations by combining multiple columns or generating statistical plots like box plots and histograms.

**A simple line chart showing Close price:**

```python
# Importing necessary libraries

import matplotlib.pyplot as plt

# Plotting the 'Close' column with specified figure size and title
df['Close'].plot(figsize=(12, 6), title='Closing Price (2017 - 2022)')

# Displaying the plot
plt.show()
```

A more advanced chart combining `Close Price` and `Volume`:

```python
import matplotlib.pyplot as plt

# Plotting the Closing Price
ax1 = df['Close'].plot(figsize=(12, 6), title='S&P Closing Price | 2017 - 2022')

# Creating a twin axis for Volume
ax2 = ax1.twinx()
df['Volume'].plot(ax=ax2, color='red', ylim=[df['Volume'].min(), df['Volume'].max() * 5])

# Adding legend
ax1.figure.legend(["Close", "Volume"], loc='upper left')

plt.show()
```

**Using subplots**

```python
# Create subplots for 'Close' price and 'Volume'
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 6))

# Plot 'Close' price with blue color
df['Close'].plot(ax=axes[0], title='S&P Closing Price', legend=True, color='blue')

# Plot 'Volume' with green color
df['Volume'].plot(ax=axes[1], title='Volume', legend=True, color='green')

# Show the plot
plt.tight_layout()
plt.show()
```

**Histogram**

```python
# Plot a histogram of 'Volume'

df['Volume'].plot(kind='hist')

# Add a title to the plot
plt.title('Histogram of Volume')

# Show the plot
plt.show()
```

**A box plot:**

```python
import matplotlib.pyplot as plt

# Create a box plot for the 'Volume' column
df['Volume'].plot(kind='box', vert=False)

# Add a title to the plot
plt.title('Box Plot of Volume')

# Show the plot
plt.show()
```

### 3. Data Wrangling

Data wrangling, often referred to as data munging, is a crucial aspect of data analysis and involves the process of cleaning, transforming, and enriching raw data into a more suitable format for analysis.

In this phase, you'll typically perform tasks such as:

- **Data Cleaning**: Detecting and handling missing or erroneous data, removing duplicates, and correcting inconsistencies.
- **Data Transformation**: Reshaping data structures, converting data types, and standardizing data formats.
- **Data Enrichment**: Adding additional information to the dataset through merging with other datasets or generating derived features.

Data wrangling ensures that the data is in a clean, consistent, and usable format, laying the foundation for accurate analysis and insights extraction.

In the context of pandas, you'll leverage its powerful functionalities to efficiently perform these data wrangling tasks, making your analysis process smoother and more effective.


**Close price and it's SMA (Simple Moving Average):**

```python
df['Close SMA'] = df['Close'].rolling(60).mean()
```

```python
df[['Close', 'Close SMA']].tail(10)
```

```python
# Calculate lower and upper bands
df['Lower Band'] = df['Close SMA'] - (2 * df['Close'].rolling(60).std())
df['Upper Band'] = df['Close SMA'] + (2 * df['Close'].rolling(60).std())
```

```python
# Plotting close price and simple moving average
ax = df[['Close', 'Close SMA']].plot(figsize=(14, 7), title='Close Price & its SMA')

# Display the plot
plt.show()
```

```python
df['Lower Band'] = df['Close SMA'] - (2 * df['Close'].rolling(60).std())
df['Upper Band'] = df['Close SMA'] + (2 * df['Close'].rolling(60).std())
df[['Close', 'Close SMA', 'Lower Band', 'Upper Band']].tail()
```
